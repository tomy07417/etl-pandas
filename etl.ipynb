## Valore nulos
Hay que revisar las tablas categories y orders

* categories: Al faltarle a todas las filas los valores de la columna 'parent_category_id' da la pauta de que ninguna tiene alguna categoría padre por lo que esa columna no proporciona informacion alguna. De querer establecer alguna categoría padre hay que hacerlo a mano.
* orders:
    - En cuanto al promotion_id el significado de tener el valor NaN da a entender que no tiene ninguna promoción esa order por lo que se lo puede reemplazar con un 0 ya que las promotion_id empiezan desde el id 1.
    - En cuanto a 'notes' el valor Nan refleja que no existe ninguna nota sobre esa orden se puede agregar el el string "-" para simbolizar que no hay.

    En ambos casos se puede dejar el valor NaN pero hay que considerarlos para futuras consultas.

```python
df_categories.drop(columns=['parent_category_id'], inplace=True)
```

## Preguntas de negocio

1) Top 5 clientes por gasto total
2) Producto más vendido
3) Evolución mensual de ventas

```python
# Top 5 clientes por gasto total
df_orders.head()
df_orders.groupby('customer_id')['total_amount'].sum().sort_values(ascending=False).reset_index().head(5)
```

```python
# Producto más vendido
df_order_items.head()
df_order_items.groupby('product_id')['quantity'].sum().sort_values(ascending=False).reset_index().head(5)
```

```python
# Evolución mensual de ventas
df_orders.head()

sales_evolution = df_orders.groupby('order_date')['total_amount'].sum().reset_index()
sales_evolution['year'] = sales_evolution['order_date'].dt.year
sales_evolution['month'] = sales_evolution['order_date'].dt.month

sales_evolution = sales_evolution.groupby(['year', 'month'])['total_amount'].sum().reset_index()
sales_evolution['date'] = pd.to_datetime(sales_evolution[['year', 'month']].assign(DAY=1))
```

```python
x = sales_evolution['date']
y = sales_evolution['total_amount']

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(x, y, marker='o')
```
